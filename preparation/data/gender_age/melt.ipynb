```python
import pandas as pd
import numpy as np
import re
```

```python
# Old format used in previous datasets
data_path="/Users/rony/Projects/Health_Geo_app/build_database/tables/2014/"
file_name="{dp}gender_age.csv".format(dp=data_path)
ds_2014=pd.read_csv(file_name)
remove=['Unnamed: 0','POSTCODE','CCG_CODE', 'NHSE_AREA_TEAM_CODE','NHSE_REGION_CODE', 'Total_All', 'Total_Male', 'Total_Female']
ds_2014=ds_2014.drop(remove, axis=1)

# New format used in monthly datasets
data_path="/Users/rony/Projects/Health_Geo_app/build_database/gender_age/monthly_data/april-2017.csv"
monthly_ex=pd.read_csv(data_path)
monthly_ex=monthly_ex[monthly_ex['AGE_GROUP_5']!='ALL']
remove=['POSTCODE','End_date']
monthly_ex=monthly_ex.drop(remove, axis=1)
```

```python
# Melt the annual data files to the new format which reflect the query UI in terms of separte gender and age
ga_melt=pd.melt(frame=ds_2014, id_vars='Practice Code', value_name='NUMBER_OF_PATIENTS')
ga_melt['SEX'] = ga_melt['variable'].map(lambda x:x.split('_', 1)[0])
ga_melt['AGE_GROUP_5'] = ga_melt['variable'].map(lambda x:x.split('_', 1)[1])
del ga_melt['variable']
ga_melt['SEX']=[i.upper() for i in list (ga_melt['SEX'])]
ga_melt=ga_melt[['Practice Code', 'SEX','AGE_GROUP_5','NUMBER_OF_PATIENTS']]
ga_melt.head()
```
