```python
import pandas as pd
import re
data_path='bnf_codes_sections.txt'
bnf_codes_sections=open(data_path).read()
bnf_codes_sections=re.sub('https://openprescribing.net/bnf/|\*|Feedback|\
University of Oxford. Funded by WEAHSN and The Health Foundation','',bnf_codes_sections).split('\n')

#bs_lines=[l.lstrip().rstrip() for l in bnf_codes_sections if (l and ("<" not in l))]
#bs_lines=['0'+re.sub('\.','0',l) for l in bnf_codes_sections]
bs_lines = ['0'+re.sub('\.','0',l).rstrip().lstrip() for l in bnf_codes_sections]
bs_lines = [l for l in bs_lines if not re.findall('^0<',l)]
bs_lines = [re.sub('<\d{1,}\/>','',l).rstrip().lstrip() for l in bs_lines][:-1]
bs_lines
```

```python
results_path='/Users/rony/Projects/Health_Geo_app/dashboard/lists/bnf_codes_sections.txt'

with open (results_path,'w') as f:
    for l in bs_lines:
        f.write ('{l}\n'.format(l=l))
```
